```python
import numpy as np
from matplotlib import cm

jan_mean = jan_df["CO2"].mean()
feb_mean = feb_df["CO2"].mean()
mar_mean = mar_df["CO2"].mean()
apr_mean = apr_df["CO2"].mean()
may_mean = may_df["CO2"].mean()
jun_mean = jun_df["CO2"].mean()
jul_mean = jul_df["CO2"].mean()
aug_mean = aug_df["CO2"].mean()
sep_mean = sep_df["CO2"].mean()
oct_mean = oct_df["CO2"].mean()
nov_mean = nov_df["CO2"].mean()
dec_mean = dec_df["CO2"].mean()

jan_max = jan_df["CO2"].max()
jan_min = jan_df["CO2"].min()
feb_max = feb_df["CO2"].max()
feb_min = feb_df["CO2"].min()
mar_max = mar_df["CO2"].max()
mar_min = mar_df["CO2"].min()
apr_max = apr_df["CO2"].max()
apr_min = apr_df["CO2"].min()
may_max = may_df["CO2"].max()
may_min = may_df["CO2"].min()
jun_max = jun_df["CO2"].max()
jun_min = jun_df["CO2"].min()
jul_max = jul_df["CO2"].max()
jul_min = jul_df["CO2"].min()
aug_max = aug_df["CO2"].max()
aug_min = aug_df["CO2"].min()
sep_max = sep_df["CO2"].max()
sep_min = sep_df["CO2"].min()
oct_max = oct_df["CO2"].max()
oct_min = oct_df["CO2"].min()
nov_max = nov_df["CO2"].max()
nov_min = nov_df["CO2"].min()
dec_max = dec_df["CO2"].max()
dec_min = dec_df["CO2"].min()

months = [jan_df["CO2"], feb_df["CO2"], mar_df["CO2"], apr_df["CO2"], may_df["CO2"], jun_df["CO2"], jul_df["CO2"], aug_df["CO2"], sep_df["CO2"], oct_df["CO2"], nov_df["CO2"], dec_df["CO2"]]

means = [jan_mean, feb_mean, mar_mean, apr_mean, may_mean, jun_mean, jul_mean, aug_mean, sep_mean, oct_mean, nov_mean, dec_mean]
maxes = [jan_max, feb_max, mar_max, apr_max, may_max, jun_max, jul_max, aug_max, sep_max, oct_max, nov_max, dec_max]
mins = [jan_min, feb_min, mar_min, apr_min, may_min, jun_min, jul_min, aug_min, sep_min, oct_min, nov_min, dec_min]

means = np.array(means)
maxes = np.array(maxes)
mins = np.array(mins)

colormap = cm.get_cmap('viridis')

month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.figure(figsize=(12, 6))

# bars = plt.bar(range(1, 13), means, color='skyblue', alpha=0.7)
for i, (month_mean, month_data) in enumerate(zip(means, [jan_df["CO2"], feb_df["CO2"], mar_df["CO2"], apr_df["CO2"], may_df["CO2"], jun_df["CO2"], jul_df["CO2"], aug_df["CO2"], sep_df["CO2"], oct_df["CO2"], nov_df["CO2"], dec_df["CO2"]]), start=1):
    colors = colormap((month_mean - min(means)) / (max(means) - min(means)))
    plt.scatter(np.full_like(month_data, i), month_data, color=colors, alpha=0.7, s=20)  # Adjust the 's' parameter for dot size
   
plt.errorbar(range(1, 13), means, yerr=[means - mins, maxes - means], fmt='none', ecolor='gray', elinewidth=1, capsize=5)

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Atmospheric CO2 Concentration (ppm)')
plt.title('Spread of Atmospheric CO2 by Month')

plt.xticks(range(1, 13), month_names, rotation=45, ha='right')

plt.ylim(300, 430)

# Show the plot

sm = plt.cm.ScalarMappable(cmap=colormap)
sm.set_array(means)
cbar = plt.colorbar(sm)
cbar.set_label('Mean Atmospheric CO2 Concentration (1958 - 2023)')

plt.show()
```
